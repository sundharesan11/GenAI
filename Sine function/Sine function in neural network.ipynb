```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 100000 samples over 0 to 2pi
```

```python
num_samples = 100000
x = np.linspace(0, 2 * np.pi, num_samples)
y = np.sin(x)
noise = np.random.normal(0, 0.1, num_samples)  # Adding Gaussian noise with mean 0 and standard deviation 0.1
y_noisy = y + noise

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Data', alpha=0.7)
plt.plot(x, y_noisy, label='Noisy Sine Data', alpha=0.7)
plt.title('Sine Data with Added Noise')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the neural network model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Convert data to TensorFlow tensors
x_tensor = tf.convert_to_tensor(x.reshape(-1, 1), dtype=tf.float32)
y_noisy_tensor = tf.convert_to_tensor(y_noisy.reshape(-1, 1), dtype=tf.float32)

# Train the neural network
history = model.fit(x_tensor, y_noisy_tensor, epochs=10, batch_size=32, verbose=1)

# Make predictions for various input values
x_test = np.linspace(0, 2 * np.pi, 500)
predictions = model.predict(x_test.reshape(-1, 1))

# Plot the predictions
plt.figure(figsize=(10, 6))
plt.plot(x, y_noisy, label='Original Data with Noise', alpha=0.7)
plt.plot(x_test, predictions, label='Predicted Data', alpha=0.7)
plt.title('Neural Network Prediction for Noisy Sine Function')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
pi_value = np.array([np.pi])
three_pi_value = np.array([3 * np.pi])
```

```python
m = model.predict(pi_value.reshape(-1, 1))
n = model.predict(three_pi_value.reshape(-1, 1))
```

```python
print("sin(pi):", m)
print("sin(3*pi):", n)
```
